# Initial data analysis

The Olympics data will be analysed to find out what would be the most useful angle to take when seeking insights from this data.

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandasql import sqldf

import warnings
warnings.filterwarnings("ignore")
```

```python
events = pd.read_csv("SportsStats/athlete_events.csv")
events.head()
```

```python
events.shape
```

```python
events.info()
```

```python
#find % nulls in columns > 0% null
cols = events.columns
null_columns = []
for i in cols:
    if events[i].isnull().sum() > 0:
        null_columns.append(i)
        print('Column {0} consists of {1}% null values'.format(i, ((events[i].isnull().sum()/len(events))*100).round(3)))
```

```python
#nulls for medals are presumably for those who did not get a medal so fill with 0
events['Medal'] = events['Medal'].fillna(0)
events.head()
```

```python
#certain sports require height and weight measurements and others do not
#need to see distribution of nulls
sports = events['Sport'].unique()
null_cols = ['Age','Height','Weight']

for i in sports:
    tmp = events[events.Sport == i]
    prop_null_age = tmp['Age'].isnull().sum()/len(tmp)*100
    prop_null_h = tmp['Height'].isnull().sum()/len(tmp)*100
    prop_null_w = tmp['Weight'].isnull().sum()/len(tmp)*100
    print('Percentage of {} that is null (of total number of entries of {}) :'.format(i, len(tmp)))
    print('    Age:', prop_null_age.round())
    print('    Height:', prop_null_h.round())
    print('    Weight:', prop_null_w.round())
```

From the above, the sport with the most entries and with comparatively few missing entries is Athletics. Therefore, this will be the focus.

```python
athletics = events[events.Sport == 'Athletics']
athletics.head()
```

Missing values will be filled in with the median values of athletes in the same event, year and gender. Although this does not guarantee accuracy in the values, it is likely that all athletes of a given event and gender will be similar for any given year. Null values, especially for medals, may negatively impact analysis. Median is used to prevent outliers impacting the value.

```python
athletics["Height"] = athletics.groupby(["Event", "Year", "Sex"])["Height"].transform(lambda x: x.fillna(x.median()))
athletics["Weight"] = athletics.groupby(["Event", "Year", "Sex"])["Weight"].transform(lambda x: x.fillna(x.median()))
athletics["Age"] = athletics.groupby(["Event", "Year", "Sex"])["Age"].transform(lambda x: x.fillna(x.median()))
```

```python
athletics.isnull().sum()
```

Still some nulls - need to find out where these are.

```python
null_ath_w = athletics[athletics.Weight.isnull()]
null_ath_w
```

```python
null_ath_w.Event.nunique()
```

All remaining null entries occur within 4 events, and each event is confined to a single year (1896, 1900, 10904), and are all male (most likely as very few women competed in these early years). As they were not filled in with a median value, this suggests all values for each event in a given year are missing so these can be removed entirely.
