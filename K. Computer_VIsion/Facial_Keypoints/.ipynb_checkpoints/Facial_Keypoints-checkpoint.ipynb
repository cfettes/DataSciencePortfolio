## Facial keypoints detection

This task will involve creating a facial keypoint detector based on CNN regressor. This was completed as part of the Computer Vision course in the Advanced Machine Learning Specialisation on Coursera.

![title](example.png)

### Load and preprocess data

File `gt.csv` contains header and ground truth points for every image. It has 29 columns. First column is a filename and next 28 columns are `x` and `y` coordinates for 14 facepoints. The following preprocessing will be conducted:
1. Scale all images to resolution $100 \times 100$ pixels.
2. Scale all coordinates to range $[-0.5; 0.5]$. To obtain that, divide all x's by width (or number of columns) of image, and divide all y's by height (or number of rows) of image and subtract 0.5 from all values.

```python
from numpy import array, zeros
from os.path import join
from skimage.color import gray2rgb
from skimage.io import imread
from skimage.transform import resize
import pandas as pd
import numpy as np
import os

file = pd.read_csv('data/gt.csv')
face_points = file[file.columns[1:]]

def load_imgs_and_keypoints(dirname='data'):
    #load images and points
    data = pd.read_csv(dirname + '/gt.csv')
    N = data.shape[0]
    points_arr = np.array(data[data.columns[1:]].values, dtype=np.float32)
    points = np.zeros_like(points_arr, dtype=np.float32)
    imgs = np.zeros((N, 100, 100, 3)) #fetching a size of 100 * 100 for image scaling
    #scaling process
    for i in range(N):
        img = imread(dirname + '/images/' + data.filename[i])
        w = img.shape[1]
        h = img.shape[0]
        points[i,:] = points_arr[i,:] / np.tile([w,h], 14) - 0.5
        img = resize(img, (100,100, 3))
        imgs[i,:,:,:] = img
    
    return imgs, points

imgs, points = load_imgs_and_keypoints()
```

### Visualize data

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

def visualize_points(img, points):
    #function to obtain image and normalised coordinates and visualises points on image
    fig, ax = plt.subplots()
    ax.imshow(img)
    for i in range(len(points) // 2):
        plt.plot((points[i*2] + 0.5)*100, (points[i*2 + 1] + 0.5)*100, 'o', color='r')
        
visualize_points(imgs[1], points[1])
```

### Train/validation split

```python
from sklearn.model_selection import train_test_split
imgs_train, imgs_val, points_train, points_val = train_test_split(imgs, points, test_size=0.1)
```

### Simple data augmentation

For better training simple data augmentation will be used, which will involve flipping an image and points. The function flip_img flips an image and its points. Coordinates must be mirrored and corresponding points swapped on the left and right sides of the face.

For correct NN training, the augmented photos should have the facial keypoints listed in the same order. The 14 keypoints are present in the original dataset in the below order. When the image is flipped, the x-coordinate of the point need to be inverted, as well as placed on the opposite side of the face.

0: left eyebrow, outer

1: left eyebrow, inner

2: right eyebrow, inner

3: right eyebrow, outer

4: left eye, outer

5: left eye, center

6: left eye, inner

7: right eye, inner

8: right eye, center

9: right eye, outer

10: nose center

11: mouth, left

12: mouth, center

13: mouth, right

```python
def flip_img(img, points):
    img_flip = img[:, ::-1]
    points_flip = np.zeros(points.shape)
    flip_indices = [3,2,1,0,9,8,7,6,5,4,10,13,12,11]
    for i in range(14):
        points_flip[2*i] = -points[2*flip_indices[i]]
        points_flip[2*i+1] = points[2*flip_indices[i]+1]
    return img_flip, points_flip

f_img, f_points = flip_img(imgs[0], points[0])
visualize_points(f_img, f_points)
```

```python
visualize_points(imgs[0], points[0])
```

The training sample will now be augmented by applying flip to every image in the training sample. As a result two arrays will be obtained: `aug_imgs_train` and `aug_points_train` which contain original images and points along with flipped ones.

```python
def augmentSample(imgs_train, points_train):
    aug_imgs = []
    aug_points = []
    for img, points in zip(imgs_train, points_train):
        f_img, f_points = flip_img(img, points)
        aug_imgs.append(img)
        aug_imgs.append(f_img)
        aug_points.append(points)
        aug_points.append(f_points)
    return np.asarray(aug_imgs), np.asarray(aug_points)

aug_imgs_train, aug_points_train = augmentSample(imgs_train, points_train)
```

### Network architecture and training

A neural network regressor will now be defined. It will have 28 outputs, 2 numbers per point. It contains 3 (`Conv2D` + `MaxPooling2D`) pairs, then `Flatten` and 3 `Dense` layers. LeakyReLU activations are used, along with `Dropout` to prevent overfitting.

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.layers.advanced_activations import LeakyReLU
from keras.initializers import glorot_uniform


model = Sequential()

model.add(Conv2D(32, (3,3), input_shape=(100,100,3), padding="same",kernel_initializer=glorot_uniform()))
model.add(LeakyReLU(alpha=0.1))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(64, (3,3), padding="same",kernel_initializer=glorot_uniform()))
model.add(LeakyReLU(alpha=0.1))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(128, (3,3), padding="same",kernel_initializer=glorot_uniform()))
model.add(LeakyReLU(alpha=0.1))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))  

model.add(Flatten())
model.add(Dense(256,kernel_initializer=glorot_uniform()))
model.add(LeakyReLU(alpha=0.1))

model.add(Dropout(0.5)) 

model.add(Dense(64,kernel_initializer=glorot_uniform()))
model.add(LeakyReLU(alpha=0))

model.add(Dense(28,kernel_initializer=glorot_uniform()))
```

```python
model.summary()
```

```python
from keras.callbacks import ModelCheckpoint
from keras.optimizers import SGD, Adam

#compile model with adam optimiser and mse as loss as training a regressor
model.compile(loss='mse', optimizer=Adam(), metrics=['mse'])
```

```python
BATCH_SIZE = 128
EPOCHS = 50
```

```python
model.fit(
    aug_imgs_train, aug_points_train,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    validation_data=(imgs_val, points_val),
    shuffle=True,
    verbose=1
)
```
