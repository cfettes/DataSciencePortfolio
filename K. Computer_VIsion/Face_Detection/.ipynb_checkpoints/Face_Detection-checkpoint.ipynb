# Face Detection

In this notebook, a deep face detector will be created. This was completed as part of, and adapted from, the Computer Vision course within the Advanced Machine Learning specialisation on Coursera.

For this task, the FDDB dataset was used (http://vis-www.cs.umass.edu/fddb/). Processing of the dataset consists of:

1. Extracting bboxes from dataset. The original dataset contains ellipsis to bound faces, defined by [here](http://vis-www.cs.umass.edu/fddb/samples/), which are not very useful for basic neural network learning.
2. Removing images with big and small faces on one shoot.
3. Re-sizing images to bounding boxes (bboxes), which have the same size 32 +/- pixels.

In the resulting processed image dataset, each image in train, validation and test datasets have shape (176, 176, 3), but part of this image has a black background, with the image aligned at the top left corner.

Bounding boxes define the face in the image and consists of 5 integer numbers: [image_index, min_row, min_col, max_row, max_col]. Bounding box width and height are 32 +/- 8 pixels wide.

`train_bboxes` and `val_bboxes` is a list of bboxes.

`train_shapes` and `val_shapes` is a list of interesting image shapes.

The data can be downloaded from [here](https://github.com/vslutov/face-detection/releases/tag/v1.3). the original_data.zip contains the unprocessed data, whereas data.zip contains preprocessed data.

```python
image_count = 0

def fold(n_fold):
    global image_count
    fnames, bboxes = [], []

    with open("data/FDDB-folds/FDDB-fold-{n_fold:02d}-ellipseList.txt".format(n_fold=n_fold), "r") as fin:
        fin = iter(fin)
        
        try:
            while True:
                fnames.append(next(fin).strip())
                shape = cv.imread("data/originalPics/" + fnames[-1] + ".jpg").shape[:2]
                
                count = int(next(fin))
                
                for i in range(count):
                    
                    a, b, phi, center_x, center_y, _1 = (float(c) for c in next(fin).split())
                    t_x = np.arctan2(-b * np.tan(phi), a )
                    x_diff = np.abs(a * np.cos(t_x) * np.cos(phi) - b * np.sin(t_x) * np.sin(phi))
                    t_y = np.arctan2(b, a * np.tan(phi))
                    y_diff = np.abs(b * np.sin(t_y) * np.cos(phi) + a * np.cos(t_y) * np.sin(phi))

                    

                    bbox = [np.floor(center_y - y_diff), np.floor(center_x - x_diff), np.ceil(center_y + y_diff), np.ceil(center_x + x_diff)]
                    bbox = [max((int(c), 0)) for c in bbox]
                    bbox[::2] = (min((c, shape[0])) for c in bbox[::2])
                    bbox[1::2] = (min((c, shape[1])) for c in bbox[1::2])
                    bbox = [image_count, *bbox, *shape]
                    
                    bboxes.append(bbox)
                
                image_count += 1
        except StopIteration:
            pass
        
        return fnames, bboxes
```

```python
convert_scales = []

for image_index in set(bboxes[:, 0]):
    image_bboxes = bboxes[bboxes[:, 0] == image_index]
    bbox_sizes = image_bboxes[:, (3, 4)] - image_bboxes[:, (1, 2)]
    avg_size = bbox_sizes.mean()
    rescale = 32 / avg_size
    
    converted_bbox_sizes = bbox_sizes * rescale
    converted_image_size = image_bboxes[0, -2:] * rescale
    
    TR = 8
    if (converted_bbox_sizes.min() >= 32 - TR and
        converted_bbox_sizes.max() <= 32 + TR and
        converted_image_size.min() >= 40 and
        converted_image_size.max() <= 176
       ):
        convert_scales.append([image_index, rescale])
```

## Prepare data

For learning, the positive and negative samples are extracted from image. Positive and negative samples counts should be similar. Every sample should have same size.

Every image can contain multiple faces, so all faces need to be extracted from every image, and cropped to `SAMPLE_SHAPE`. This set of extracted images are named `positive`.

Then the `negative` set will be extracted. These images will also have `SAMPLE_SHAPE` size. 

```python
SAMPLE_SHAPE = (32, 32, 3)
```

```python
from scores import iou_score #https://en.wikipedia.org/wiki/Jaccard_index

def is_negative_bbox(new_bbox, true_bboxes, eps=1e-1):
    """Check if new bbox not in true bbox list; bbox is 4 ints [min_row, min_col, max_row, max_col] without image index."""
    for bbox in true_bboxes:
        if iou_score(new_bbox, bbox) >= eps:
            return False
    return True
```

```python
import skimage.transform

def gen_negative_bbox(image_shape, bbox_size, true_bboxes):
    """Generate negative bbox for image."""
    height, width = bbox_size
    min_row = np.random.choice(image_shape[0] - height)
    max_row = min_row + height
    min_col = np.random.choice(image_shape[1] - width)
    max_col = min_col + width
    return [min_row, min_col, max_row, max_col]

def get_positive_negative(images, true_bboxes, image_shapes, negative_bbox_count=None):
    """Retrieve positive and negative samples from image."""
    positive = []
    negative = []
    image_count = image_shapes.shape[0]
    
    if negative_bbox_count is None:
        negative_bbox_count = len(true_bboxes)
    
    #extract negative samples only from part [0:image_shape[0], 0:image_shape[1]]
    for true_bbox in true_bboxes:        
        image_index = true_bbox[0]
        pos_img = images[image_index][true_bbox[1]:true_bbox[1]+32, true_bbox[2]:true_bbox[2]+32, :]
        positive.append(pos_img)
        
    for i in range(negative_bbox_count):
        image_index = np.random.choice(len(images))
        image_shape = image_shapes[image_index]
        image_true_bboxes = true_bboxes[true_bboxes[:, 0] == image_index, 1:]
        for j in range(100):
            bbox_size = (32, 32)
            new_bbox = gen_negative_bbox(image_shape, bbox_size, image_true_bboxes)
            if is_negative_bbox(new_bbox, image_true_bboxes):
                neg_img = images[image_index][new_bbox[0]:new_bbox[2], new_bbox[1]:new_bbox[3]]
                negative.append(neg_img)
                break
   
    return positive, negative
```

```python
def get_samples(images, true_bboxes, image_shapes):
    """Useful samples for learning.
    X - positive and negative samples.
    Y - one hot encoded list of zeros and ones. One is positive marker.
    """
    positive, negative = get_positive_negative(images=images, true_bboxes=true_bboxes, 
                                               image_shapes=image_shapes)
    X = positive
    Y = [[0, 1]] * len(positive)
    
    X.extend(negative)
    Y.extend([[1, 0]] * len(negative))
    
    return np.array(X), np.array(Y)
```

```python
import os.path
from keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau, LearningRateScheduler

def fit(model, datagen, X_train, Y_train, X_val, Y_val, model_name=None, output_dir="data/checkpoints", class_weight=None, epochs=50, lr=0.001, verbose=False):
    
    if verbose:
        model.summary()

    model.compile(optimizer=Adam(lr=lr),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    
    history = model.fit_generator(datagen.flow(X_train, Y_train, batch_size=BATCH_SIZE),
                                  validation_data=(datagen.standardize(X_val), Y_val),
                                  epochs=epochs, steps_per_epoch=len(X_train) // BATCH_SIZE,
                                  callbacks=[ModelCheckpoint(os.path.join(output_dir, "{model_name}").format(model_name=model_name) + "-{epoch:02d}-{val_loss:.2f}.hdf5", save_best_only=True),
                                            ] if model_name is not None else [],
                                  class_weight=class_weight,
                      
                                 )  
    
    #summarise history for accuracy
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'val'], loc='upper left')
    plt.show()
    #summarise history for loss
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'val'], loc='upper left')
    plt.show()
```

```python
from keras.models import Model, Sequential
from keras.layers import Flatten, Dense, Activation, Input, Dropout, Activation, BatchNormalization
from keras.layers import Conv2D, MaxPooling2D, GlobalAveragePooling2D

def generate_model(sample_shape):
    #classification model
    x = inputs = Input(shape=sample_shape)
    x = Conv2D(32, (5, 5), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Conv2D(32, (5, 5), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Flatten()(x)
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.25)(x)
    x = Dense(64, activation='relu')(x)
    x = Dropout(0.25)(x)

    predictions = Dense(2, activation='softmax')(x)
    return Model(inputs=inputs, outputs=predictions)

model = generate_model(SAMPLE_SHAPE)
```

## Detection

The classification architecture can now be used for detection. Converting the classification architecture to a fully convolution neural network (FCNN) returns a heatmap of activation.

### Detector model or sliding window

Now the fully-connected layers will be replaced with $1 \times 1$ convolution layers. Every fully connected layer performs the operation $f(Wx + b)$, where $f(\cdot)$ is the nonlinear activation function, $x$ is the layer input, and $W$ and $b$ are the layer weights. This operation can be emulated with $1 \times 1$ convolution with the activation function $f(\cdot)$, that perform exactly the same operation $f(Wx + b)$.

As there is a `Flatten` layer with $n \times k$ input size before the fully connected layer, convolution will have the same $n \times k$ input size. As there were multiple fully connected layers, these are replaced with a convolution layers sequence.

After replacing fully connected layers with convolution layers, the result is a fully convolution network. When input shape is equal to input size of the previous network, the output will have the size $1 \times 1$. But if input shape was increased, output shape automatically will be increased. For example, if convolution step of previous network strides 4 pixels, increased input size with 100 pixels along all axes generates increased output size with 25 values along all axes.

In total:
1. $1 \times 1$ convolution layer is equivalent to a fully connected layer.
2. $1 \times 1$ convolution layers can be used to get activation map of classification network in "sliding window" manner.

The last fully connected layer with softmax activation is replaced by a convolution layer with linear activation.

```python
#FCNN
IMAGE_SHAPE = (176, 176, 3)

def generate_fcnn_model(image_shape):
    """After model compilation input size cannot be changed.
    So, a function needs to be created that has the ability to change size later.
    """
    x = inputs = Input(image_shape)

    x = Conv2D(32, (5, 5), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Conv2D(32, (5, 5), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2))(x)

    x = Conv2D(128, (8, 8), activation='relu')(x)
    x = Dropout(0.25)(x)
    x = Conv2D(64, (1, 1), activation='relu')(x)
    x = Dropout(0.25)(x)

    predictions = Conv2D(2, (1, 1), activation='linear')(x)
    return Model(inputs=inputs, outputs=predictions)

fcnn_model = generate_fcnn_model(IMAGE_SHAPE)
```

```python
#function that copies weights without modifications from classification model to fully convolutional model
def copy_weights(base_model, fcnn_model):
    """Set FCNN weights from base model.
    """
    
    new_fcnn_weights = []
    prev_fcnn_weights = fcnn_model.get_weights()
    prev_base_weights = base_model.get_weights()
            
    for i in range(len(prev_fcnn_weights)):    
        print(i,prev_fcnn_weights[i].shape,prev_base_weights[i].shape)
        new_fcnn_weights.append(prev_base_weights[i].reshape(prev_fcnn_weights[i].shape))
        
    fcnn_model.set_weights(new_fcnn_weights)

copy_weights(base_model=model, fcnn_model=fcnn_model)
```

### Detector

First detector part gets bboxes and decision function. A greater decision function indicates better detector confidence.

The below functions return pred_bboxes and decision_function:

- `pred bboxes` is list of 5 int tuples like `true bboxes`: `[image_index, min_row, min_col, max_row, max_col]`.
- `decision function` is confidence of detector for every pred bbox: list of float values, `len(decision function) == len(pred bboxes)` 

```python
from skimage import filters

def get_peaks(image, min_dist, threshold_rel=None, num_peaks=-1):

    shape_0 = image.shape
    image = filters.gaussian(image, sigma=1)
    if (image.shape != shape_0):
        image = transform.resize(image, shape_0)

    image_flat = []
    max_val = image[0][0]
    min_val = image[0][0]
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            entry = (image[i][j], i, j)
            image_flat.append(entry)
            max_val = max(max_val, image[i][j])
            min_val = min(min_val, image[i][j])

    image_flat = reversed(sorted(image_flat, key=lambda x: x[0]))
    threshold_abs = threshold_rel*max_val if threshold_rel is not None else min_val

    image_tags = np.ones(image.shape)
    detected_peaks = []
    for val, r, c in image_flat:
        if val < threshold_abs:
            break
        if image_tags[r][c] == 1:
            detected_peaks.append([r, c])
            up_row = max(0, r-min_dist)
            dw_row = min(image.shape[0]-1, r+min_dist)
            le_col = max(0, c-min_dist)
            ri_col = min(image.shape[1]-1, c+min_dist)
            image_tags[up_row:dw_row+1, le_col:ri_col+1] = 0
        if num_peaks > 0 and len(detected_peaks) >= num_peaks:
            break

    return detected_peaks
```

```python
#detection
from skimage.feature import peak_local_max
from skimage.color import rgb2gray

def get_bboxes_and_decision_function(fcnn_model, images, image_shapes):      
    cropped_images = np.array([transform.resize(image, IMAGE_SHAPE, mode="reflect")  if image.shape != IMAGE_SHAPE else image for image in images])
    pred_bboxes, decision_function = [], []

    #predict
    predictions = fcnn_model.predict(datagen.standardize(np.asarray(cropped_images,dtype='float')))

    for i in range(len(predictions)):

        pred_img = predictions[i]
        hm_scaled = transform.resize(pred_img[:, :, 1], (IMAGE_SHAPE[0] - SAMPLE_SHAPE[0] + 1, IMAGE_SHAPE[1] - SAMPLE_SHAPE[1] + 1))
        min_dist = int(min(SAMPLE_SHAPE[0], SAMPLE_SHAPE[1]))

        raw_peaks = get_peaks(hm_scaled, min_dist=min_dist, threshold_rel=0.0)

        dtc_bboxes = []
        dtc_functs = []
      
        for p in raw_peaks:
            r = p[0]
            c = p[1]
            delta_h = SAMPLE_SHAPE[0]-1
            delta_w = SAMPLE_SHAPE[1]-1
            bbox = [None]*5
            bbox[0] = i
            bbox[1] = r
            bbox[2] = c
            bbox[3] = min(IMAGE_SHAPE[0]-1, r + delta_h)
            bbox[4] = min(IMAGE_SHAPE[1]-1, c + delta_w)
            funct = hm_scaled[r][c]

            #only add bbox if contained in original shape
            if r < image_shapes[i][0] and c < image_shapes[i][1]:
                dtc_bboxes.append(bbox)
                dtc_functs.append(funct)

        pred_bboxes.extend(dtc_bboxes)
        decision_function.extend(dtc_functs)

    #scale/normalise decision function 
    if decision_function:
            dec_min = min(decision_function)
            dec_max = max(decision_function)
            decision_function = [100.0*d for d in decision_function]

    return pred_bboxes, decision_function
```

## Detector score

The below function creates the [precision and recall](https://en.wikipedia.org/wiki/Precision_and_recall) graph.

The function `best_match` is utilised by this function to extract matching prediction and ground truth, false positive and false negative samples.

```python
from scores import best_match
from graph import plot_precision_recall

def precision_recall_curve(pred_bboxes, true_bboxes, decision_function):
    precision, recall, thresholds = [], [], []
    
    threshold = min(decision_function) - 1
    max_th = max(decision_function) + 1
    num_steps = 100
    th_step = (max_th - threshold)/num_steps

    sorted_boxes = [[x]+y for y,x in sorted(zip(pred_bboxes, decision_function), key = lambda pair:pair[1])]
    
    for step in range(num_steps):        
        pred_bboxes_th = [x[1:] for x in sorted_boxes if x[0] > threshold]
        if len(pred_bboxes_th) > 0:
            matched, false_negative, false_positive = best_match(pred_bboxes_th, true_bboxes, decision_function)
        else:
            break
        
        prec = len(matched)/(len(matched)+len(false_positive))
        rec = len(matched)/(len(matched)+len(false_negative))
        
        thresholds.append(threshold)
        recall.append(rec)
        precision.append(prec)
        threshold += th_step
    
    return precision, recall, thresholds
```

```python
precision, recall, thresholds = precision_recall_curve(pred_bboxes = pred_bboxes, true_bboxes = val_bboxes, decision_function = decision_function)
plot_precision_recall(precision = precision, recall = recall)
```

### Threshold

Next step in the detector creation is to select a threshold for the decision_function. Every possible threshold presents point on the recall-precision graph.

The threshold selected is `recall = 0.85`.

```python
def get_threshold(thresholds, recall):
    return thresholds[np.argmax(np.asarray(recall) <= 0.85)]

THRESHOLD = get_threshold(thresholds, recall)
```

```python
def detect(fcnn_model, images, image_shapes, threshold, return_decision = True):
    """Get bboxes with decision_function not less then threshold."""
    pred_bboxes, decision_function = get_bboxes_and_decision_function(fcnn_model, images, image_shapes)   
    result, result_decision = [], []
    
    for i in range(len(pred_bboxes)):
        if decision_function[i] >= threshold:
            result.append(pred_bboxes[i])
            result_decision.append(decision_function[i])
    
    if return_decision:
        return result, result_decision
    else:
        return result
```

```python
pred_bboxes, decision_function = detect(fcnn_model = fcnn_model, images = val_images, image_shapes = val_shapes, threshold = THRESHOLD, return_decision = True)

visualize_bboxes(images = val_images,
                 pred_bboxes = pred_bboxes,
                 true_bboxes = val_bboxes,
                 decision_function = decision_function
                )

precision, recall, thresholds = precision_recall_curve(pred_bboxes = pred_bboxes, true_bboxes = val_bboxes, decision_function = decision_function)
plot_precision_recall(precision = precision, recall = recall)
```

## Hard negative mining

[Hard negative mining](https://www.reddit.com/r/computervision/comments/2ggc5l/what_is_hard_negative_mining_and_how_is_it/) can upgrade the score of the model. A hard negative is when a falsely detected patch is taken, and a negative example is explicitly created out of that patch, and that negative is added to the training set. When the classifier is retrained, it should perform better with this extra knowledge, and not make as many false positives.

```python
def hard_negative(train_images, train_shapes, train_bboxes, X_val, Y_val, base_model, fcnn_model):
    #pred_bboxes, decision_function = get_bboxes_and_decision_function(fcnn_model=fcnn_model, images=original_images,  image_shapes=original_shapes)
    pred_bboxes, decision_function = detect(fcnn_model=fcnn_model, images=train_images, image_shapes=train_shapes, threshold=THRESHOLD, return_decision=True)

    pos, fneg, fpos = best_match(pred_bboxes, train_bboxes, decision_function)

    X_train_fpos = []
    Y_train_fpos = []

    for idx in fpos:
        bbox = pred_bboxes[idx]  
        img = train_images[bbox[0]]
        sample = img[bbox[1]:bbox[3]+1, bbox[2]:bbox[4]+1, :]
        X_train_fpos.append(sample)
        Y_train_fpos.append([1, 0])

    X_train_fpos = np.stack(X_train_fpos, axis=0 )
    X_train_fpos = np.array(X_train_fpos)
    Y_train_fpos = np.array(Y_train_fpos)

    X_train, Y_train = get_samples(train_images, train_bboxes, train_shapes)
    
    X_train = np.concatenate((X_train, X_train_fpos))
    Y_train = np.concatenate((Y_train, Y_train_fpos))

    fit(model_name="lenet-hnm", model=base_model, datagen=datagen, X_train=X_train, X_val=X_val, Y_train=Y_train, Y_val=Y_val)
```

```python
hard_negative(train_images=train_images, train_shapes=train_shapes, train_bboxes=train_bboxes, X_val=X_val, Y_val=Y_val, base_model=model, fcnn_model=fcnn_model)
```

### Multi scale detector

The multi-scale detector will be created and tested with [pyramid representation][pyramid].
[pyramid]: https://en.wikipedia.org/wiki/Pyramid_(image_processing)

1. Resize images to predefined scales.
2. Run detector with different scales.
3. Apply non-maximum supression to detection on different scales.

References:
1. [E. H. Adelson,C. H. Anderson, J. R. Bergen, P. J. Burt, J. M. Ogden: Pyramid methods in image processing](http://persci.mit.edu/pub_pdfs/RCA84.pdf)
2. [PETER J. BURT, EDWARD H. ADELSON: The Laplacian Pyramid as a Compact Image Code](http://persci.mit.edu/pub_pdfs/pyramid83.pdf)

```python
from skimage.util import pad

def rescale_pad(image, ratio, shape):
    ratio = min(ratio, 1.0)
    ar = float(image.shape[0]/image.shape[1])
    h = int(shape[0]*ratio) if ar > 1.0 else int(shape[1]*ratio*ar)
    w = int(h/ar)
    img = transform.resize(image, (h, w, image.shape[2]), mode='constant')
    img = pad(img, ((0, shape[0]-h), (0, shape[1]-w), (0,0)))
    img_shape = (h, w, image.shape[2])
    return img, img_shape
```

```python
from scores import iou_score

eps = 0.05

def nms(pred_bbox, decision_function):
    idx_list = reversed(np.argsort(decision_function))

    pred_bbox_new, decision_function_new = [], []
    for idx in idx_list:
        is_max = True
        bbox = pred_bbox[idx]
        for bbox_other in pred_bbox_new:
            if bbox[0] == bbox_other[0] and iou_score(bbox[1:], bbox_other[1:]) >= eps:
                is_max = False
                break
        if is_max:
            pred_bbox_new.append(pred_bbox[idx])
            decision_function_new.append(decision_function[idx])

    return pred_bbox_new, decision_function_new
```

```python
def multiscale_detector(fcnn_model, images, image_shapes):
    pred_bboxes = []
    decision_function = []

    for i, img in enumerate(images):
        #search for best scale of detection
        for r in [1.0, 0.75, 0.5, 0.25]:
            n_img, n_shape = rescale_pad(img, r, IMAGE_SHAPE)
            pb, df = get_bboxes_and_decision_function(fcnn_model=fcnn_model, images=[n_img], image_shapes=[n_shape])

            max_scale = float(img.shape[0]/n_shape[0])
            for j in range(len(pb)):
                bbox = pb[j]
                pred_bbox = [None]*len(bbox)
                pred_bbox[0] = i
                pred_bbox[1] = int(bbox[1]*max_scale)
                pred_bbox[2] = int(bbox[2]*max_scale)
                pred_bbox[3] = int(bbox[3]*max_scale)
                pred_bbox[4] = int(bbox[4]*max_scale)
                pred_bboxes.append(pred_bbox)
                decision_function.append(df[j])

    pred_bboxes, decision_function = nms(pred_bboxes, decision_function)

    return pred_bboxes, decision_function
```

```python
pred_bboxes, decision_function = multiscale_detector(fcnn_model = fcnn_model, images = original_images, image_shapes = original_shapes)

visualize_bboxes(images = original_images,
                 pred_bboxes = pred_bboxes,
                 true_bboxes = original_bboxes,
                 decision_function = decision_function
                )

precision, recall, threshold = precision_recall_curve(pred_bboxes = pred_bboxes, true_bboxes = original_bboxes, decision_function = decision_function)
plot_precision_recall(precision = precision, recall = recall)
```

This has clearly hugely improved the performance of the model on the unprocessed dataset - it is still not great, but it is definitely signficiantly better.
